# Landmarks in pixel space

```python
import imageio
import numpy as np
%matplotlib notebook
#%matplotlib inline
import matplotlib.pyplot as plt
```

```python
img_01 = imageio.imread('frame_01.png')
img_15 = imageio.imread('frame_15.png')

fig, ax = plt.subplots()
im = ax.imshow(img_01, interpolation='none')
plt.axis('off')
plt.show()
```

```python
lm_coords = dict(
    lm1=(297, 409),
    lm2=(297, 247),
    lm3=(297, 134),
    lm4=(297, 63),
    lm5=(161, 104),
    lm6=(425, 104),
    lm7=(143, 305),
    lm10=(223, 303),
    lm13=(360, 301),
    lm16=(435, 297),
    lm19=(234, 437),
    lm20=(366, 437),
    lm24=(204, 552),
    lm26=(297, 510),
    lm28=(399, 543),
    lm30=(297, 578),
    lm38=(305, 678),
    lm39=(156, 578),
    lm40=(448, 578),
    lm41=(117, 431),
    lm42=(104, 318),
    lm43=(489, 424),
    lm44=(489, 301),
    lm46=(158, 229),
    lm49=(420, 212)
)

lm_coords2 = dict(
    lm1=(297, 409),
    lm2=(297, 247),
    lm3=(297, 134),
    lm4=(297, 63),
    lm5=(161, 104),
    lm6=(425, 104),
    lm7=(143, 305),
    lm10=(223, 303),
    lm13=(360, 301),
    lm16=(435, 297),
    lm19=(234, 437),
    lm20=(366, 437),
    lm24=(204, 562),
    lm26=(297, 510),
    lm28=(399, 563),
    lm30=(297, 620),
    lm38=(305, 708),
    lm39=(156, 578),
    lm40=(448, 578),
    lm41=(117, 431),
    lm42=(104, 318),
    lm43=(489, 424),
    lm44=(489, 301),
    lm46=(158, 229),
    lm49=(420, 212)
)
```

```python
%matplotlib inline
fig, ax = plt.subplots(ncols=3, figsize=(16, 12))
im = ax[0].imshow(img_01, interpolation='none')
for lm, coord in lm_coords.items():
    ax[0].plot([coord[0]] * 2, [coord[1]] * 2, marker='o', ms=5, color='b')

    for lm2, coord2 in lm_coords.items():
        
        if lm == lm2:
            continue
        
        dist = np.sqrt((coord[0] - coord2[0]) ** 2 + (coord[1] - coord2[1]) ** 2)
        if dist < 250:
            ax[0].plot([coord[0], coord2[0]], [coord[1], coord2[1]], c='b', lw=0.5)    

    ax[0].axis("off")
    
im = ax[1].imshow(img_15, interpolation='none')
for lm, coord in lm_coords2.items():
    ax[1].plot([coord[0]] * 2, [coord[1]] * 2, marker='o', ms=5, color='b')

    for lm2, coord2 in lm_coords.items():
        
        if lm == lm2:
            continue
        
        dist = np.sqrt((coord[0] - coord2[0]) ** 2 + (coord[1] - coord2[1]) ** 2)
        if dist < 250:
            ax[1].plot([coord[0], coord2[0]], [coord[1], coord2[1]], c='b', lw=0.5)    

    ax[1].axis("off")

im = ax[2].imshow(img_15, interpolation='none')
for lm, coord in lm_coords2.items():

    dist = (lm_coords[lm][0] - lm_coords2[lm][0]) ** 2 + (lm_coords[lm][1] - lm_coords2[lm][1]) ** 2
    if dist > 0:
        ax[2].plot([coord[0]] * 2, [coord[1]] * 2, marker='o', ms=5, color='b')

        for lm2, coord2 in lm_coords.items():

            if lm == lm2:
                continue
            dist = np.sqrt((coord[0] - coord2[0]) ** 2 + (coord[1] - coord2[1]) ** 2)
            if dist < 250:
                ax[2].plot([coord2[0]] * 2, [coord2[1]] * 2, marker='o', ms=5, color='b')
                ax[2].plot([coord[0], coord2[0]], [coord[1], coord2[1]], c='b', lw=0.5)   

    ax[2].axis("off")

plt.savefig('landmarks_diff.png', transparent=True)
plt.show()
```
